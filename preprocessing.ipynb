```python
def inserting_images(path, lis):
    print(lis)
    li = [os.getcwd()+"\\Corel10000"+"\\"+image for image in os.listdir("./Corel10000") if image.endswith(".jpg")]
    #print(li[10])
    for i in range(100):
        for j in range(len(li)):
            #print(path+"\\"+"train"+"\\"+str(i)+"\\", len(path+"\\"+"train"+"\\"+str(i)+"\\"))
            if li[j][42:44] == lis[i] and len(os.listdir(path+"\\"+"train"+"\\"+str(i)+"\\")) < 89:
                print("train ", li[j][42:44], "lis ", lis[i])
                #print("train :-", li[j][42:])
                #print("Now it is 90 percent...", li[j][-6:-4])
                img = cv2.imread(li[j], cv2.IMREAD_COLOR)
                if img is not None:
                    #print(path+"\\"+"train"+"\\"+str(i)+"\\"+"00"+f"{li[42:-4]}.jpg")
                    cv2.imwrite(path+"\\"+"train"+"\\"+str(i)+"\\"+"00"+f"{li[j][42:-4]}.jpg", img)
                else:
                    pass
                    print("None Image....")
                os.remove(li[j])
            elif li[j][42:44] == lis[i] and len(os.listdir(path+"\\"+"test"+"\\"+str(i)+"\\")) < 10:
                print("train ", li[j][42:44], "lis ", lis[i])
                #print("train :-", li[j][42:])
                #print("Now it is 90 percent...", li[j][-6:-4])
                img = cv2.imread(li[j], cv2.IMREAD_COLOR)
                if img is not None:
                    #print(path+"\\"+"train"+"\\"+str(i)+"\\"+"00"+f"{li[42:-4]}.jpg")
                    cv2.imwrite(path+"\\"+"test"+"\\"+str(i)+"\\"+"00"+f"{li[j][42:-4]}.jpg", img)
                else:
                    pass
                    print("None Image....")
                os.remove(li[j])
            elif li[j][42:44] == lis[i] and len(os.listdir(path+"\\"+"val"+"\\"+str(i)+"\\")) < 2:
                print("train ", li[j][42:44], "lis ", lis[i])
                #print("train :-", li[j][42:])
                #print("Now it is 90 percent...", li[j][-6:-4])
                img = cv2.imread(li[j], cv2.IMREAD_COLOR)
                if img is not None:
                    #print(path+"\\"+"train"+"\\"+str(i)+"\\"+"00"+f"{li[42:-4]}.jpg")
                    cv2.imwrite(path+"\\"+"val"+"\\"+str(i)+"\\"+"00"+f"{li[j][42:-4]}.jpg", img)
                else:
                    pass
                    print("None Image....")
                os.remove(li[j])
            else:
                pass
                

inserting_images(r"C:\Users\SUPERMAN\Desktop\hand\image_Sep", lis)
```
